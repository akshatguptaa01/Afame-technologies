```python
import pandas as pd
import numpy as np
```

```python
hrData = pd.read_csv('HR Data.csv')
```

```python
columns_to_exclude = ['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours']

# Exclude the unnecessary columns
hrData_filtered = hrData.drop(columns=columns_to_exclude)
```

```python
hrData_filtered
```

```python
# Dictionary to map old column names to new names
new_column_names = {
    'Age': 'Age',
    'Attrition': 'Attrition',
    'BusinessTravel': 'Business_Travel',
    'DailyRate': 'Daily_Rate',
    'Department': 'Department',
    'DistanceFromHome': 'Distance_From_Home',
    'Education': 'Education',
    'EducationField': 'Education_Field',
    'EmployeeCount': 'Employee_Count',
    'EmployeeNumber': 'Employee_Number',
    'EnvironmentSatisfaction': 'Environment_Satisfaction',
    'Gender': 'Gender',
    'HourlyRate': 'Hourly_Rate',
    'JobInvolvement': 'Job_Involvement',
    'JobLevel': 'Job_Level',
    'JobRole': 'Job_Role',
    'JobSatisfaction': 'Job_Satisfaction',
    'MaritalStatus': 'Marital_Status',
    'MonthlyIncome': 'Monthly_Income',
    'MonthlyRate': 'Monthly_Rate',
    'NumCompaniesWorked': 'Num_Companies_Worked',
    'Over18': 'Over_18',
    'OverTime': 'Over_Time',
    'PercentSalaryHike': 'Percent_Salary_Hike',
    'PerformanceRating': 'Performance_Rating',
    'RelationshipSatisfaction': 'Relationship_Satisfaction',
    'StandardHours': 'Standard_Hours',
    'StockOptionLevel': 'Stock_Option_Level',
    'TotalWorkingYears': 'Total_Working_Years',
    'TrainingTimesLastYear': 'Training_Times_Last_Year',
    'WorkLifeBalance': 'Work_Life_Balance',
    'YearsAtCompany': 'Years_At_Company',
    'YearsInCurrentRole': 'Years_In_Current_Role',
    'YearsSinceLastPromotion': 'Years_Since_Last_Promotion',
    'YearsWithCurrManager': 'Years_With_Current_Manager'
}

# Rename the columns
hrData_renamed = hrData.rename(columns=new_column_names)
```

```python
hrData_renamed
```

```python
#remove the duplicate values
hrData_renamed.drop_duplicates()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

# 1. Handling Missing Values
# Replace missing values in numeric columns with the mean
numeric_columns = ['Age', 'Daily_Rate', 'Monthly_Income']
imputer = SimpleImputer(strategy='mean')
hrData_renamed.loc[:, numeric_columns] = imputer.fit_transform(hrData_renamed[numeric_columns])

# 2. Removing Outliers
# Remove outliers in 'Monthly_Income' column using IQR
Q1 = hrData_renamed['Monthly_Income'].quantile(0.25)
Q3 = hrData_renamed['Monthly_Income'].quantile(0.75)
IQR = Q3 - Q1
hrData_renamed = hrData_renamed[~((hrData_renamed['Monthly_Income'] < (Q1 - 1.5 * IQR)) | (hrData_renamed['Monthly_Income'] > (Q3 + 1.5 * IQR)))]

# 3. Standardization
# Standardize 'Age' column
scaler = StandardScaler()
hrData_renamed.loc[:, 'Age'] = scaler.fit_transform(hrData_renamed[['Age']])
```

```python
hrData_renamed
```

```python
#eliminate the nan values
hrData_renamed.fillna(0, inplace=True)
```
